```python
train_data = [
    ("What is the price of 10 shares of Apple on May 5?", {
        "entities": [(21, 23, "QUANTITY"), (34, 39, "ORG"), (43, 49, "DATE")]
    }),
    ("How much for 5 Tesla stocks today at 3 PM?", {
        "entities": [(13, 14, "QUANTITY"), (15, 20, "ORG"), (28, 33, "DATE"), (37, 41, "TIME")]
    }),
    ("Show me the price of 3 Amazon shares at 2:30 PM", {
        "entities": [(24, 25, "QUANTITY"), (26, 32, "ORG"), (36, 43, "TIME")]
    }),
    ("I want to buy 8 Microsoft stocks on Tuesday", {
        "entities": [(16, 17, "QUANTITY"), (18, 27, "ORG"), (33, 40, "DATE")]
    }),
    ("Give me the rate for 12 Netflix shares next Monday", {
        "entities": [(23, 25, "QUANTITY"), (26, 33, "ORG"), (34, 45, "DATE")]
    }),
    ("What's the cost of 6 Nvidia shares on 12th June?", {
        "entities": [(20, 21, "QUANTITY"), (22, 28, "ORG"), (32, 43, "DATE")]
    }),
    ("Can I purchase 15 Meta stocks after 1 PM?", {
        "entities": [(17, 19, "QUANTITY"), (20, 24, "ORG"), (32, 37, "TIME")]
    }),
    ("Is 4 Alphabet stock available before 10 AM?", {
        "entities": [(3, 4, "QUANTITY"), (5, 13, "ORG"), (29, 36, "TIME")]
    }),
    ("Price check for 7 AMD shares this Friday", {
        "entities": [(17, 18, "QUANTITY"), (19, 22, "ORG"), (29, 40, "DATE")]
    }),
    ("What’s the price of 2 Uber stocks on 5th May?", {
        "entities": [(30, 31, "QUANTITY"), (32, 36, "ORG"), (40, 48, "DATE")]
    }),
    ("Check if 9 Zoom shares are rising today", {
        "entities": [(9, 10, "QUANTITY"), (11, 15, "ORG"), (35, 40, "DATE")]
    }),
    ("Show me 20 Intel shares at 11:00 AM", {
        "entities": [(8, 10, "QUANTITY"), (11, 16, "ORG"), (20, 28, "TIME")]
    }),
    ("How much are 10 Adobe stocks worth next Friday?", {
        "entities": [(13, 15, "QUANTITY"), (16, 21, "ORG"), (34, 45, "DATE")]
    }),
    ("Can I buy 5 Salesforce shares after 4 PM?", {
        "entities": [(11, 12, "QUANTITY"), (13, 23, "ORG"), (30, 35, "TIME")]
    }),
    ("Tell me the value of 25 PayPal shares this evening", {
        "entities": [(22, 24, "QUANTITY"), (25, 31, "ORG"), (32, 44, "TIME")]
    }),
    ("I need info for 3 Oracle shares before market closes", {
        "entities": [(17, 18, "QUANTITY"), (19, 25, "ORG"), (33, 50, "TIME")]
    }),
    ("What’s the trend for 6 Boeing stocks on Thursday?", {
        "entities": [(21, 22, "QUANTITY"), (23, 29, "ORG"), (33, 42, "DATE")]
    }),
    ("Get me 7 Disney shares details now", {
        "entities": [(8, 9, "QUANTITY"), (10, 16, "ORG")]
    }),
    ("Track 2 Ford stocks as of 9:30 AM", {
        "entities": [(6, 7, "QUANTITY"), (8, 12, "ORG"), (20, 28, "TIME")]
    }),
    ("I want updates on 9 Starbucks shares by tonight", {
        "entities": [(19, 20, "QUANTITY"), (21, 30, "ORG"), (34, 42, "TIME")]
    }),
    ("Is 3 Costco stock trending tomorrow?", {
        "entities": [(3, 4, "QUANTITY"), (5, 11, "ORG"), (29, 37, "DATE")]
    }),
    ("How is 1 Walmart share performing at 10:15 AM?", {
        "entities": [(7, 8, "QUANTITY"), (9, 16, "ORG"), (32, 41, "TIME")]
    }),
    ("Can I get quotes for 6 Nike shares this morning?", {
        "entities": [(22, 23, "QUANTITY"), (24, 28, "ORG"), (29, 42, "TIME")]
    }),
    ("Give me a report on 4 Pepsi stocks for next week", {
        "entities": [(21, 22, "QUANTITY"), (23, 28, "ORG"), (33, 43, "DATE")]
    }),
    ("Check the movement of 10 Chevron shares on Monday morning", {
        "entities": [(23, 25, "QUANTITY"), (26, 33, "ORG"), (37, 53, "DATE")]
    }),
    ("What is the price of 10 Volvoworks shares today?", {
        "entities": [(21, 23, "QUANTITY"), (24, 34, "ORG"), (42, 47, "DATE")]
    }),
    ("How much does 5 Volvoworks shares cost today at $120?", {
        "entities": [(16, 17, "QUANTITY"), (18, 28, "ORG"), (46, 50, "PRICE"), (41, 45, "DATE")]
    }),
    ("I want to buy 20 Volvoworks stocks for USD 100", {
        "entities": [(14, 16, "QUANTITY"), (17, 27, "ORG"), (37, 45, "PRICE")]
    }),
    ("Price of 3 Volvoworks shares on April 30 is $90", {
        "entities": [(9, 10, "QUANTITY"), (11, 21, "ORG"), (29, 38, "DATE"), (42, 45, "PRICE")]
    }),
    ("Check 15 Volvoworks shares at 10:30 AM for $150", {
        "entities": [(6, 8, "QUANTITY"), (9, 19, "ORG"), (27, 35, "TIME"), (41, 45, "PRICE")]
    }),
    ("I can't log into my Volvoworks account since yesterday", {
        "entities": [(21, 31, "ORG"), (39, 48, "DATE")]
    }),
    ("Please reset my password for Volvoworks", {
        "entities": [(32, 42, "ORG")]
    }),
    ("I need help with the premium report from Volvoworks", {
        "entities": [(21, 35, "PRODUCT"), (41, 51, "ORG")]
    }),
    ("My premium dashboard subscription from Volvoworks costs $49", {
        "entities": [(3, 29, "PRODUCT"), (35, 45, "ORG"), (53, 56, "PRICE")]
    }),
    ("Volvoworks customer support called me at 2 PM", {
        "entities": [(0, 10, "ORG"), (41, 45, "TIME")]
    }),
    ("Can you resend the 2FA link for my Volvoworks login?", {
        "entities": [(36, 46, "ORG")]
    }),
    ("I want a refund for my premium plan charged on May 2", {
        "entities": [(29, 41, "PRODUCT"), (54, 59, "DATE")]
    }),
    ("I was charged $25 on April 28 by Volvoworks", {
        "entities": [(14, 17, "PRICE"), (21, 30, "DATE"), (34, 44, "ORG")]
    }),
    ("Received my trading report from Volvoworks at 3:45 PM", {
        "entities": [(12, 27, "PRODUCT"), (33, 43, "ORG"), (47, 55, "TIME")]
    }),
    ("What is the monthly report fee on May 1? It was $30", {
        "entities": [(12, 25, "PRODUCT"), (33, 38, "DATE"), (48, 51, "PRICE")]
    }),
    ("Buy 3 trading signals from Volvoworks for $70", {
        "entities": [(4, 5, "QUANTITY"), (6, 22, "PRODUCT"), (28, 38, "ORG"), (43, 46, "PRICE")]
    }),
    ('I want to buy the apples',{
        "entities": [(18,24,"PRODUCT")]
    }),
    ('I want to buy the bananas',{
        "entities": [(18,25,"PRODUCT")]
    }),
]


nlp=spacy.load('en_core_web_md')
ner=nlp.get_pipe('ner')

# adding labels if they are missing 
for _, annotations in train_data:
    for ent in annotations['entities']:
        if ent not in ner.labels:
            ner.add_label(ent[2])
```

```python
intent = ['']
```

```python
other_pipes = [pipe for pipe,_ in nlp.pipeline if pipe != 'ner'] # pipelines apart from the ner
with nlp.disable_pipes(*other_pipes): #disabling the other pipeline
    optimizer=nlp.begin_training # begin the training 
    epochs=100
    for epoch in range(epochs):
        random.shuffle(train_data)
        losses={}
        batches=minibatch(train_data,size=2)
        for batch in batches:
            examples=[]
            for text, annotations in batch:
                doc=nlp.make_doc(text)
                example=Example.from_dict(doc,annotations)
                examples.append(example)
            nlp.update(examples,drop=0.5,losses=losses)
        print(f'Epoch{epoch+1},Losses: {losses}')
nlp.to_disk('custom_ner_model')
trained_nlp=spacy.load('custom_ner_model')
```
