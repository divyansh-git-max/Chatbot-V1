# fine-tuning

```python
from spacy.util import minibatch
from spacy.training.example import Example
import random
```

```python
train_data=[
    ("what is the price of 10 bananas ? " ,{"entities": [(21,23,"QUANTITY"),(24,31,"PRODUCT")]}),
    ("what is the price of 5 apples?", {"entities": [(21, 22, "QUANTITY"), (23, 29, "PRODUCT")]}),
    ("how much for 6 apples?", {"entities": [(13, 14, "QUANTITY"), (15, 21, "PRODUCT")]}),
    ("price of 8 apples today?", {"entities": [(10, 11, "QUANTITY"), (12, 18, "PRODUCT")]}),
    ("tell me the cost of 12 apples", {"entities": [(21, 23, "QUANTITY"), (24, 30, "PRODUCT")]}),
    ("I need price info for 4 apples", {"entities": [(23, 24, "QUANTITY"), (25, 31, "PRODUCT")]}),
    ("how much do 12 eggs cost?", {"entities": [(13, 15, "QUANTITY"), (16, 20, "PRODUCT")]}),
    ("rate of 6 eggs please", {"entities": [(8, 9, "QUANTITY"), (10, 14, "PRODUCT")]}),
    ("what's the price of 10 eggs?", {"entities": [(22, 24, "QUANTITY"), (25, 29, "PRODUCT")]}),
    ("cost for 18 eggs today", {"entities": [(9, 11, "QUANTITY"), (12, 16, "PRODUCT")]}),
    ("price check for 30 eggs", {"entities": [(17, 19, "QUANTITY"), (20, 24, "PRODUCT")]}),
    ("What is the price of 2 laptops?", {"entities": [(21, 22, "QUANTITY"), (23, 30, "PRODUCT")]}),
    ("How much does 1 mic cost?", {"entities": [(16, 17, "QUANTITY"), (18, 21, "PRODUCT")]}),
    ("Tell me the cost of 3 headphones.", {"entities": [(21, 22, "QUANTITY"), (23, 33, "PRODUCT")]}),
    ("I want to buy 5 pens.", {"entities": [(16, 17, "QUANTITY"), (18, 22, "PRODUCT")]}),
    ("What's the rate for 4 keyboards?", {"entities": [(22, 23, "QUANTITY"), (24, 33, "PRODUCT")]}),
    ("Can you check the price of 6 chairs?", {"entities": [(29, 30, "QUANTITY"), (31, 37, "PRODUCT")]}),
    ("How much are 2 books?", {"entities": [(13, 14, "QUANTITY"), (15, 20, "PRODUCT")]}),
    ("Give me the cost for 3 bottles.", {"entities": [(22, 23, "QUANTITY"), (24, 31, "PRODUCT")]}),
    ("Price of 1 phone?", {"entities": [(9, 10, "QUANTITY"), (11, 16, "PRODUCT")]}),
    ("I need to know the price of 8 desks.", {"entities": [(30, 31, "QUANTITY"), (32, 37, "PRODUCT")]}),
    ("What is the cost of 2 monitors?", {"entities": [(20, 21, "QUANTITY"), (22, 30, "PRODUCT")]}),
    ("How much does 7 mugs cost?", {"entities": [(16, 17, "QUANTITY"), (18, 22, "PRODUCT")]}),
    ("Tell me the price of 9 t-shirts.", {"entities": [(23, 24, "QUANTITY"), (25, 33, "PRODUCT")]}),
    ("I want to purchase 6 tables.", {"entities": [(21, 22, "QUANTITY"), (23, 29, "PRODUCT")]}),
    ("What's the price of 4 cameras?", {"entities": [(22, 23, "QUANTITY"), (24, 31, "PRODUCT")]})
]
nlp=spacy.load('en_core_web_md')
ner=nlp.get_pipe('ner')

# adding labels if they are missing 
for _, annotations in train_data:
    for ent in annotations['entities']:
        if ent not in ner.labels:
            ner.add_label(ent[2])
```

```python
nlp.pipe_labels
```

```python
nlp.pipe_names
```

```python
other_pipes = [pipe for pipe,_ in nlp.pipeline if pipe != 'ner'] # pipelines apart from the ner
with nlp.disable_pipes(*other_pipes): #disabling the other pipeline
    optimizer=nlp.begin_training # begin the training 
    epochs=50
    for epoch in range(epochs):
        random.shuffle(train_data)
        losses={}
        batches=minibatch(train_data,size=2)
        for batch in batches:
            examples=[]
            for text, annotations in batch:
                doc=nlp.make_doc(text)
                example=Example.from_dict(doc,annotations)
                examples.append(example)
            nlp.update(examples,drop=0.5,losses=losses)
        print(f'Epoch{epoch+1},Losses: {losses}')
nlp.to_disk('custom_ner_model')
trained_nlp=spacy.load('custom_ner_model')
```

```python
test_text= [
    'How much for 3 Oranges?',
    "What's the price of 4 Camera?"
]
for text in test_text:
    doc=trained_nlp(text)
    print(f'text{text}')
    print('entities',[ (ent.text,ent.label_) for ent in doc.ents])
    print()
```
